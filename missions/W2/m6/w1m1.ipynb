```python
plt.figure(figsize=(10,8))

features = df.columns[1:]

for i, col in enumerate(features):
    plt.subplot(4,3,i+1)
    df[f'{col}'].hist()
    plt.title(f'{col}')

plt.suptitle("Histograms of Variables")
plt.tight_layout()

plt.show()
```

```python
corr_matrix = df.iloc[:, 1:].corr()
corr_matrix
```

```python
import numpy as np

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

high_corr = upper.stack().reset_index()

high_corr.columns = ['Variable 1', 'Variable 2', 'Correlation']

result1 = high_corr[high_corr['Correlation'].abs() >= 0.7].sort_values(by='Correlation', ascending=False)

result2 = high_corr[high_corr['Correlation'].abs() >= 0.7].sort_values(
    by='Correlation', 
    ascending=False, 
    key=lambda x: x.abs()
)

print(result2.to_string(index=False))
```

```python
df = pd.read_csv('mtcars.csv')
df = df.rename(columns={'Unnamed: 0': 'name'})

plt.figure(figsize=(12, 8))
plt.scatter(df['hp'], df['mpg'], alpha=0.6)


for i, txt in enumerate(df['name']):
    plt.annotate(txt, 
                 (df['hp'][i], df['mpg'][i]), 
                 xytext=(5, 5),            
                 textcoords='offset points', 
                 fontsize=9)


plt.xlabel('hp')
plt.ylabel('mpg')
plt.title('MPG vs HP with Car Names')
plt.grid(True, linestyle='--', alpha=0.5)


plt.show()
```
