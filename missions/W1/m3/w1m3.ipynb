```python
from bs4 import BeautifulSoup
import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
header = {'User-Agent': 'Chrome/143.0.0.0'}
response = requests.get(url, headers=header)
html_content = response.text
html_content
```

```python
soup = BeautifulSoup(html_content,"html.parser")
gdp_table = soup.find("table", class_="wikitable")
rows = gdp_table.find_all("tr")
rows
```

```python
import re
data=[]
for row in rows:
    cols = [re.sub(r'\[.*?\]', '',ele.text).strip() for ele in row.find_all(["td", "th"])]
    data.append(cols)

df = pd.DataFrame(data[1:], columns=data[0])
df
```

```python
df = df.iloc[1:,0:2]
def convert_unit(val):
    num_str = re.sub(r'[^\d.]', '', str(val))
    if not num_str:
        return None
    return round(float(num_str)/1000, 2)

df.iloc[:,1] = df.iloc[:,1].apply(convert_unit)
    
df=df.dropna()
df
```

```python
filtered_df = df[df.iloc[:,1] >= 100]
print(filtered_df)
```
